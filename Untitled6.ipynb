```python

# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

print(tf.__version__)
```

```python

# Chargement de MNIST
from tensorflow.keras.datasets import mnist  # (attention: dans le TP c'était 'minst', ici correct: 'mnist')
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python

# Distribution des classes (train)
import pandas as pd
cls_counts = pd.Series(y_train).value_counts().sort_index()
print(cls_counts)

# Affichage sous forme de barres
plt.figure(figsize=(6,4))
cls_counts.plot(kind='bar')
plt.title("Distribution des classes - Entraînement")
plt.xlabel("Classe")
plt.ylabel("Fréquence")
plt.show()
```

```python

# Échantillon d'images 3x5
plt.figure(figsize=(8,4))
for i in range(15):
    plt.subplot(3,5,i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(int(y_train[i]))
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python

# Normalisation & reshaping vers (28,28,1)
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0
x_train = np.expand_dims(x_train, -1)
x_test  = np.expand_dims(x_test, -1)
input_shape = x_train.shape[1:]
input_shape
```

```python
model = keras.Sequential([
    layers.Conv2D(32, (5,5), activation='relu', input_shape=input_shape),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(100, activation='relu'),
    layers.Dense(10, activation='softmax')
])
model.summary()
```

```python

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

es = keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)
history = model.fit(
    x_train, y_train,
    validation_split=0.1,
    epochs=10,
    batch_size=128,
    callbacks=[es],
    verbose=2
)
```

```python

# Courbes d'apprentissage
plt.figure(figsize=(6,4))
plt.plot(history.history['accuracy'], label='acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.title("Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title("Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print("Test accuracy:", test_acc)

y_pred = model.predict(x_test, verbose=0).argmax(axis=1)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cm)
disp.plot(values_format='d')
plt.title("Matrice de confusion - Keras")
plt.show()

print(classification_report(y_test, y_pred))
```
